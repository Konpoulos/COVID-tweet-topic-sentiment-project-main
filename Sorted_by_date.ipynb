# Libraries

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

# Files

```python
csv_dir = "Csv/"
data_dir = "Data/"
```

```python
files = [file for file in os.listdir(csv_dir) if file.startswith("t") and not file.endswith("october28.csv")]
# files[:5]
```

# Set months and days to loop over

```python
months = ["march", "april", "may", "june", 
          "july", "august", "september", "october", "november", "december", "january", "february"] # from march (start dataset)
```

```python
# added a "_" so we can separate 1 from 11 (both start with a "1")
days = [str(number+1)+"_" if number+1<10 else str(number+1) for number in list(range(31))] 
# days
```

# Loop over months --> days

```python
sorted_dates_list = []
for month in months:
    month_files = [file for file in files if file[10:].startswith(month)]
    l_month = len(month)
    for day in days:
        try: #try except needed because not all days exist (e.g. no 31 of april)
            next_file = [month_file for month_file in month_files if month_file[10+l_month:10+l_month+2] == day][0]
            sorted_dates_list.append(next_file)
        except:
            print(f"no {month}_{day}")
```

I see we dont have march 29, but there indeed is no file on ieee. Rest makes sense

# Combine all tweet data

```python
all_together = pd.DataFrame()
for file in sorted_dates_list:
    df = pd.read_csv(csv_dir+file, index_col=0)
    df = df[["date", "id", "tweet", "sentiment"]]
    all_together = pd.concat([all_together, df])
all_together = all_together.reset_index(drop=True)
# all_together.to_csv(data_dir+"all_tweets_sorted.csv")
```

```python
months = ["march", "april", "may", "june", 
          "july", "august", "september", "october", "november", "december", "january", "february"] # from march (start dataset)
new_dates = all_together["date"].apply(lambda x: x.split("_")[0]) #only take first day
new_dates = new_dates.apply(lambda x: x+"_2020" if not x.startswith("jan") else x+"_2021") # add year
new_dates = list(new_dates) # make list
for i, date in enumerate(new_dates):
    for month in months: 
        l_month = len(month)
        if date.startswith(month):
            new_dates[i]=date[:l_month]+"_"+date[l_month:] #put a _ after the month
for i, date in enumerate(new_dates):
    zfilled = "_".join([date.split("_")[0], date.split("_")[1].zfill(2), date.split("_")[2]]) # split on _ and make the day zfilled (1 becomes 01)
    new_dates[i] = zfilled
new_dates = ["may_14_2020" if date == "may_14-may15_2020" else date for date in new_dates ]
all_together["date"] = new_dates
```

```python
all_together["date"] = pd.to_datetime(all_together["date"], format="%B_%d_%Y")
```
