```python
for csv in csv_list: #my_csv:
    if not os.path.isfile(csv_dir+"tweets_"+country+"_"+csv):
        tweet_data = []
        errors = 0
        current_day = pd.read_csv(csv_dir+csv, header=None)
        if len(current_day.columns) == 1:
            current_day["sentiment"] = [None]*len(current_day)
        current_day.columns = ["id", "sentiment"]
        current_day.head()
        stop = len(current_day)
        start_list = list(range(0, stop, 100))
        stop_list = [(i + 100) for i in list(range(0, stop, 100))]
        stop_list[-1]=stop
        for i in range(len(start_list)):
            current_ids = list(current_day["id"][start_list[i]: stop_list[i]])
            statuses = api.statuses_lookup(current_ids, tweet_mode = "extended")
            for i, status in enumerate(statuses):
                try:
                    if status.place.country_code == country:
                        tweet_data.append([current_day.iloc[i]["id"], 
                                           status.full_text, 
                                           current_day.iloc[i]["sentiment"],
                                           csv.split(".")[0]])
                except:
                    errors += 1
        tweet_data_df = pd.DataFrame(tweet_data, columns=["id", "tweet", "sentiment", "date"])
        tweet_data_df.to_csv(csv_dir+"tweets_"+country+"_"+csv)
        print(f"DONE ({errors} error(s)): "+csv_dir+"tweets_"+country+"_"+csv)
    else:
        print("Already DONE "+csv_dir+"tweets_"+country+"_"+csv)
print("COMPLETELY DONE")
```
